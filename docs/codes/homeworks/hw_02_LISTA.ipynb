```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

districts = pd.read_csv('districts_info.csv')
products = pd.read_csv('products_info.csv')
engagement = pd.read_csv('engagement_data/8815.csv')

print("Districts shape:", districts.shape)
print("Products shape:", products.shape)

districts.head()
```

```python
print("Duplicados en districts:", districts.duplicated().sum())
print("Duplicados en products:", products.duplicated().sum())

print("\nValores nulos por columna (districts):")
print(districts.isnull().sum())

print("\nValores nulos por columna (products):")
print(products.isnull().sum())
```

No se detectan registros duplicados en ninguno de los dos datasets, sin embargo, en districts_info existen valores nulos en la mayoría de las columnas, especialmente 'pp_total_raw' (financiamiento) y 'pct_free/reduced' (nivel socioeconómico).

Ahora verificaremos que la limpieza se haya aplicado y que no existan valores faltantes o filas vacías en los datasets principales.

```python
print("Valores nulos en districts_clean:")
print(districts_clean.isnull().sum())

print("\nValores nulos en products_clean:")
print(products_clean.isnull().sum())

print("\nTipos de datos en districts_clean:")
print(districts_clean.dtypes)
```

Lo siguiente será examinar el archivo de engagement_data correspondiente al distrito 8815. Este dataset representa la participación digital semanal de los estudiantes en 2020. Analizaremos su estructura, valores nulos y tipos de variables para preparar el análisis temporal y las comparaciones entre distritos.

```python
print("Dimensiones del dataset:", engagement.shape)
print("\nPrimeras 8 filas:")
display(engagement.head(8))

print("\nTipos de datos:")
print(engagement.dtypes)

print("\nValores nulos por columna:")
print(engagement.isnull().sum())

print("\nDescripción estadística de las columnas numéricas:")
print(engagement.describe())

print("\nValores únicos por columna:")
for col in engagement.columns:
    print(f"- {col}: {engagement[col].nunique()} únicos")
```

Lo siguiente será realizar limpieza del dataset de engagement. Se convierte la columna de fechas al tipo datetime para permitir análisis temporales, se verifican valores faltantes y se eliminan registros vacíos o inconsistentes. Además, se revisarán posibles valores atípicos en las métricas de acceso y participación.

```python
engagement_clean = engagement.copy()
engagement_clean['time'] = pd.to_datetime(engagement_clean['time'], errors='coerce')

engagement_clean = engagement_clean.dropna(subset=['time'])
print("Valores nulos después de convertir fechas:")
print(engagement_clean.isnull().sum())
print("\nRango temporal del dataset:")
print(engagement_clean['time'].min(), "→", engagement_clean['time'].max())

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(engagement_clean['pct_access'], bins=50)
plt.title('Distribución de pct_access')

plt.subplot(1,2,2)
sns.histplot(engagement_clean['engagement_index'], bins=50)
plt.title('Distribución de engagement_index')
plt.tight_layout()
plt.show()

#para valores atipicos
pct_threshold = engagement_clean['pct_access'].quantile(0.99)
eng_threshold = engagement_clean['engagement_index'].quantile(0.99)

print(f"\nPercentil 99 de pct_access: {pct_threshold:.2f}")
print(f"Percentil 99 de engagement_index: {eng_threshold:.2f}")

engagement_filtered = engagement_clean[
    (engagement_clean['pct_access'] <= pct_threshold) &
    (engagement_clean['engagement_index'] <= eng_threshold)
]

print("\nDimensiones antes del filtrado:", engagement_clean.shape)
print("Dimensiones después del filtrado:", engagement_filtered.shape)
```

Tras la conversión y limpieza de los datos de participación, se observa que el rango temporal abarca desde enero hasta diciembre de 2020. Las distribuciones de ambas métricas presentan una fuerte asimetría positiva, lo que indica que la mayoría de los distritos acceden a los recursos digitales en niveles muy bajos, mientras que unos pocos alcanzan valores excepcionalmente altos. Al eliminar los valores extremos (más allá del percentil 99), el dataset se reduce a 102,213 registros, conservando las observaciones más representativas para el análisis.

Se analizará la evolución temporal del acceso (pct_access) y del compromiso digital (engagement_index) a lo largo de 2020. Se agruparán los datos por semana y se calcularán los promedios, con el fin de visualizar las tendencias generales en la participación digital del distrito durante el año.

```python
engagement_time = engagement_clean.groupby(pd.Grouper(key='time', freq='W'))[['pct_access', 'engagement_index']].mean().reset_index()

plt.figure(figsize=(12,5))
plt.plot(engagement_time['time'], engagement_time['pct_access'], color='royalblue')
plt.title('Evolución semanal del porcentaje de acceso - 2020')
plt.xlabel('Fecha')
plt.ylabel('Porcentaje de acceso promedio')
plt.grid(True)
plt.show()

plt.figure(figsize=(12,5))
plt.plot(engagement_time['time'], engagement_time['engagement_index'], color='darkorange')
plt.title('Evolución semanal del índice de compromiso - 2020')
plt.xlabel('Fecha')
plt.ylabel('Índice de compromiso promedio')
plt.grid(True)
plt.show()

engagement_time.head()
```

Durante el año 2020 se observa un cambio drástico en el uso de plataformas digitales. A inicios del año, el nivel de acceso promedio era muy bajo, pero a partir de marzo —coincidiendo con el inicio de la pandemia y la transición a la educación remota— se registra un aumento acelerado y sostenido en la participación digital, alcanzando un punto máximo hacia mediados de año. Posteriormente, tanto el porcentaje de acceso como el índice de compromiso muestran una disminución gradual y se estabilizan en niveles más bajos, lo que sugiere una adaptación posterior de las prácticas digitales. Este patrón refleja el impacto inmediato de la crisis sanitaria sobre la adopción de herramientas tecnológicas en el sistema educativo.

En esta sección, analizaremos qué herramientas o plataformas educativas tuvieron mayor participación digital durante el año 2020. Este análisis busca identificar cuáles fueron las tecnologías más utilizadas por los estudiantes en el contexto de la educación remota, y observar cómo se distribuyó su nivel de uso a lo largo del año.

```python
product_usage = (
    engagement_filtered.groupby('lp_id')[['pct_access', 'engagement_index']]
    .mean()
    .reset_index()
)

top_products = product_usage.sort_values(by='engagement_index', ascending=False).head(10)
top_products_info = top_products.merge(products_clean, left_on='lp_id', right_on='LP ID', how='left')
top_products_info[['lp_id', 'Product Name', 'Provider/Company Name', 'Sector(s)', 'Primary Essential Function', 'pct_access', 'engagement_index']]
```

Los resultados muestran que las plataformas más utilizadas durante 2020 estuvieron dominadas por herramientas de Google, como YouTube, Google Docs y Google Sheets, todas con niveles de acceso y compromiso significativamente altos.
YouTube lidera claramente con un índice de compromiso promedio de más de 3000.

Además, se observa la presencia de plataformas complementarias como Kahoot!, Quizlet y Epic!, enfocadas en aprendizaje interactivo y recursos para estudiantes de nivel escolar, lo que sugiere un uso mixto entre herramientas de productividad y entretenimiento educativo.
Además, aparecen Instagram y Netflix, lo que podría indicar que parte del acceso digital se concentró en sitios no solamente educativos.

Siguiendo lo anterior, se analizará la evolución temporal del uso de las plataformas más populares durante 2020, observando cómo varió su porcentaje de acceso promedio a lo largo de las semanas del año.

```python
top_products = ['YouTube', 'Google Docs', 'Kahoot!', 'Quizlet', 'Epic! - Unlimited Books for Kids']

eng_top = engagement.merge(products_clean, left_on='lp_id', right_on='LP ID')
eng_top = eng_top[eng_top['Product Name'].isin(top_products)].copy()
eng_top['time'] = pd.to_datetime(eng_top['time'])
eng_top['week'] = eng_top['time'].dt.to_period('W').apply(lambda r: r.start_time)
weekly_access = eng_top.groupby(['week', 'Product Name'])['pct_access'].mean().reset_index()

plt.figure(figsize=(14,7))
sns.lineplot(data=weekly_access, x='week', y='pct_access', hue='Product Name')
plt.title('Evolución semanal del porcentaje de acceso de los principales productos - 2020')
plt.xlabel('Semana')
plt.ylabel('Porcentaje de acceso promedio')
plt.legend(title='Producto')
plt.show()

weekly_access.head()
```

El gráfico muestra la evolución semanal del porcentaje de acceso a los principales productos durante 2020. Se observa que YouTube domina ampliamente todo el período, seguido por Google Docs, mientras que Kahoot!, Quizlet y Epic! presentan niveles mucho menores. Los picos de uso más notorios ocurren entre marzo y mayo de 2020, coincidiendo con el inicio de la educación remota, y luego tienden a estabilizarse con una leve disminución hacia fin de año.

```python
merged = engagement.merge(products_clean, left_on='lp_id', right_on='LP ID', how='left')
category_summary = merged.groupby("Primary Essential Function")[["pct_access", "engagement_index"]].mean().reset_index()
category_summary = category_summary.sort_values(by="pct_access", ascending=False)
print("Resumen por tipo de función educativa:")
print(category_summary.head(10))

plt.figure(figsize=(10,5))
sns.barplot(data=category_summary, x="pct_access", y="Primary Essential Function", color="skyblue")
plt.title("Promedio de porcentaje de acceso por tipo de plataforma (2020)")
plt.xlabel("Porcentaje promedio de acceso")
plt.ylabel("Tipo de plataforma")
plt.show()

plt.figure(figsize=(10,5))
sns.barplot(data=category_summary, x="engagement_index", y="Primary Essential Function", color="lightcoral")
plt.title("Promedio de índice de compromiso por tipo de plataforma (2020)")
plt.xlabel("Índice promedio de compromiso")
plt.ylabel("Tipo de plataforma")
plt.show()
```

Durante 2020, los sistemas de gestión del aprendizaje (LMS) concentraron la mayor parte del uso y del compromiso digital, con un promedio de acceso casi tres veces superior al resto de categorías. Les siguieron las herramientas de videoconferencia y los recursos de evaluación y participación en el aula virtual, reflejando el cambio abrupto hacia la educación remota.
Por el contrario, los recursos complementarios como las plataformas de estudio, contenidos o sitios de referencia mostraron una participación mucho menor, lo que sugiere que la prioridad durante la pandemia fue mantener la continuidad educativa antes que diversificar herramientas.

Por ultimo, se analizará si existen diferencias significativas en el acceso y compromiso digital entre distritos urbanos, suburbanos, rurales, etc., y entre distintos estados. Para esto, cargaremos todos los distritos de engagement

```python
import glob

files = glob.glob("engagement_data/*.csv")

eng_list = []

for f in files:
    df = pd.read_csv(f)
    district_id = f.split("/")[-1].replace(".csv", "")
    df["district_id"] = district_id
    eng_list.append(df)

engagement_full = pd.concat(eng_list, ignore_index=True)
engagement_full["time"] = pd.to_datetime(engagement_full["time"], errors="coerce")
engagement_full["district_id"] = pd.to_numeric(engagement_full["district_id"], errors="coerce")
print("Engagement full shape:", engagement_full.shape)
engagement_full.head()
```

```python
eng_geo = engagement_full.merge(districts_clean, on="district_id", how="left")
locale_access = eng_geo.groupby('locale')[['pct_access', 'engagement_index']].mean().reset_index()
locale_access = locale_access.sort_values('pct_access', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(data=locale_access, x='locale', y='pct_access', palette='crest')
plt.title('Acceso promedio por tipo de zona (locale) - 2020')
plt.xlabel('Tipo de zona')
plt.ylabel('Porcentaje de acceso promedio')
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(data=locale_access, x='locale', y='engagement_index', palette='crest')
plt.title('Índice de compromiso promedio por tipo de zona (locale) - 2020')
plt.xlabel('Tipo de zona')
plt.ylabel('Índice de compromiso promedio')
plt.show()

state_access = eng_geo.groupby('state')[['pct_access', 'engagement_index']].mean().reset_index()
state_access = state_access.sort_values('pct_access', ascending=False)

plt.figure(figsize=(14,7))
sns.barplot(data=state_access, x='state', y='pct_access', palette='crest')
plt.title('Acceso promedio por estado - 2020')
plt.xlabel('Estado')
plt.ylabel('Porcentaje de acceso promedio')
plt.xticks(rotation=90)
plt.show()

state_access.head()
```

En general, los resultados muestran diferencias claras según el contexto geográfico. Las zonas urbanas (City) presentan los valores más altos tanto en acceso como en índice de compromiso, seguidas por los suburbios. En contraste, las áreas rurales y los pueblos muestran niveles considerablemente menores, lo que sugiere brechas digitales vinculadas al tipo de zona. La categoría “Sin dato” confirma que muchos distritos sin registro presentan menor participación promedio. A nivel estatal, también se observan variaciones importantes: estados como North Dakota, Arizona, New York y New Hampshire evidencian un mayor acceso promedio, mientras que otros, como Florida, Texas y North Carolina, se ubican en la parte baja. Esto indica que el acceso y el compromiso no solo dependen de la oferta tecnológica, sino también de condiciones estructurales y de la infraestructura educativa y digital de cada territorio.

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")

train_data.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
display(train_data.info())
print("\nNulos por columna (train):")
print(train_data.isnull().sum())
print("\nNulos por columna (test):")
print(test_data.isnull().sum())

train_data.describe(include='all').T
```

```python
train = train_data.copy()
test  = test_data.copy()

train["Title"] = train["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
test["Title"]  = test["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)

replace_titles = {
    'Lady':'Rare','Countess':'Rare','Capt':'Rare','Col':'Rare','Don':'Rare','Dr':'Rare',
    'Major':'Rare','Rev':'Rare','Sir':'Rare','Jonkheer':'Rare','Dona':'Rare'
}
for df in [train, test]:
    df['Title'] = df['Title'].replace(replace_titles)
    df['Title'] = df['Title'].replace({'Mlle':'Miss','Ms':'Miss','Mme':'Mrs'})

for df in [train, test]:
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)


train[['Name','Title','SibSp','Parch','FamilySize','IsAlone']].head(10)
```

```python
mediana_edad = train_data["Age"].median()
train_data["Age"].fillna(mediana_edad, inplace=True)
test_data["Age"].fillna(mediana_edad, inplace=True)

moda_embarked = train_data["Embarked"].mode()[0]
train_data["Embarked"].fillna(moda_embarked, inplace=True)

mediana_fare = test_data["Fare"].median()
test_data["Fare"].fillna(mediana_fare, inplace=True)

print("Nulos imputados")

def obtener_titulo(nombre):
    return nombre.split(",")[1].split(".")[0].strip()

train_data["Titulo"] = train_data["Name"].apply(obtener_titulo)
test_data["Titulo"] = test_data["Name"].apply(obtener_titulo)

titulos_raros = ['Lady','Countess','Capt','Col','Don','Dr','Major','Rev','Sir','Jonkheer','Dona']
train_data["Titulo"] = train_data["Titulo"].replace(titulos_raros, "Raro")
test_data["Titulo"] = test_data["Titulo"].replace(titulos_raros, "Raro")

print("Títulos extraídos")

train_data["TamanoFamilia"] = train_data["SibSp"] + train_data["Parch"] + 1
test_data["TamanoFamilia"] = test_data["SibSp"] + test_data["Parch"] + 1

train_data["VaSolo"] = (train_data["TamanoFamilia"] == 1).astype(int)
test_data["VaSolo"] = (test_data["TamanoFamilia"] == 1).astype(int)

print("Columnas 'TamanoFamilia' y 'VaSolo' creadas.")
train_data[["Name", "Titulo", "TamanoFamilia", "VaSolo"]].head()
```

```python
columnas_modelo = [
    "Pclass",
    "Sex",
    "Age",
    "Fare",
    "Embarked",
    "Titulo",
    "TamanoFamilia",
    "VaSolo"
]

X = pd.get_dummies(train_data[columnas_modelo])
X_test = pd.get_dummies(test_data[columnas_modelo])

X_test = X_test.reindex(columns=X.columns, fill_value=0)

y = train_data["Survived"]

modelo = RandomForestClassifier(
    n_estimators=300,
    max_depth=7,
    random_state=42
)
modelo.fit(X, y)

print("Modelo entrenado")

# Predicciones
predicciones = modelo.predict(X_test)

salida = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": predicciones
})

salida.to_csv("submission.csv", index=False)

print("submission.csv creado.")
```
